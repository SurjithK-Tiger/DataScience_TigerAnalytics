> **SalesData.xlsx dataset**

> **Question 1-6**



```python
#Importing and Reading
import pandas as pd
df = pd.read_excel("SalesData.xlsx")
df.head()
```

```python
#Question 1
df.groupby('Item')['Sale_amt'].min().reset_index()
```

```python
#Question 2
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
df['Year'] = df['OrderDate'].dt.year
df.groupby(['Year', 'Region'])['Sale_amt'].sum().reset_index()
```

```python
#Question 3
import datetime
reference_date = datetime.datetime(2025, 6, 6)

df['days_diff'] = (reference_date - df['OrderDate']).dt.days
df.head()
```

```python
#Question 4
salesmen_df = df.groupby('Manager')['SalesMan'].apply(list).reset_index()
salesmen_df.columns = ['manager', 'list_of_salesmen']
salesmen_df.head()
```

```python
#Question 5
region_sales = df.groupby('Region').agg(salesmen_count=('SalesMan', 'nunique'), total_sales=('Sale_amt', 'sum')).reset_index()
region_sales.head()
```

```python
#Question 6
total_sales_all = df['Sale_amt'].sum()
manager_sales = df.groupby('Manager')['Sale_amt'].sum().reset_index()
manager_sales['percent_sales'] = (manager_sales['Sale_amt'] / total_sales_all) * 100
manager_sales.head()
```

> **diamonds.csv dataset**

```python
#Importing and Reading
import pandas as pd
df2 = pd.read_csv("diamonds.csv")
df2.head()
```

```python
len(df2)
```

```python
#Question 11
df2.duplicated().sum()
```

```python
#Question 12
df2 = df2.dropna(subset=['carat', 'cut'])
len(df2)
```

```python
#Question 13
numeric_df = df2.select_dtypes(include=['number'])
numeric_df.head()
```

```python
#Question 14
import numpy as np
df2['volume'] = np.where(df2['depth'] > 60, df2['x'] * df2['y'] * df2['z'], 8)
df2.head()
```

```python
#Question 15
df2['price'] = df2['price'].fillna(df2['price'].mean())
```
